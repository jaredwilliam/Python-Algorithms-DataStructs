# Queue Overview

A queue is an ordered collection of items where the addition of new items happens at one end, called the "rear", and the removal of existing items occurs at the other end, commonly called the "front". 

As elements enteres the queue it starts at the rear and makes its way toward the front, waiting until that time when it is the next element to be removed. 

The most recently added item in the queue must wait at the end of the collection. The item that has been in the collection the longest is at the front. The is the ordering principle called FIFO, first in first out, or first come first served. 

The simplest example is waiting in line.
